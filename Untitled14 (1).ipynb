```python
df.columns = df.columns.str.strip()
mode_Product = df['Product'].mode()[0]
df['Product'] = df['Product'].fillna(mode_Product)
```

```python
df
```

```python
df.fillna(0)
```

```python
df.columns = df.columns.str.strip()
money_columns = [
    'Sales',
    'Profit',
    'COGS',
    'Units Sold',
    'Sale Price',
    'Discounts'
]

# Remove symbols and convert to numeric
for col in money_columns:
    df[col] = (
        df[col]
        .astype(str)                      # ensure string
        .str.replace('$', '' , regex=False)
        .str.replace(',', '', regex=False)
    )
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
scaler = StandardScaler()
kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')

# Select numerical columns for scaling and clustering
clustering_features = df[money_columns]
scaled_data = scaler.fit_transform(clustering_features)

```

```python
scaled_data = np.nan_to_num(scaled_data)
df['cluster'] = kmeans.fit_predict(scaled_data)


cluster_mean = df.groupby('cluster')['Discounts'].mean()
performance = cluster_mean.sort_values().index.tolist()
mapping = {
    performance[0]: "Low Performing",
    performance[1]: "Medium Performing",
    performance[2]: "High Performing"
}
```

```python
df['performance'] = df['cluster'].map(mapping)
```
